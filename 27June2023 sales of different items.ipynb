```python
ax1=sns.countplot(data=df, x='Age Group', hue='Gender') #remove hue and the gender column is gone 

for bars in ax1.containers: #remove this and the values above the graph is gone
    ax1.bar_label(bars)
    
#the 'hue' makes devision between genders, now we have a seggrigation between men and women of same age group. 
#wthout this there wont be a segrigation.
```

# creating a bar graph stating the amount spent by each age group

```python
ax2= df1.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount', data=ax2)
```

```python
ax2
```

```python
ax3= df1.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=True)
ax3
#When as_index is set to True, the grouped column(s) become the index, and when it is set to False, 
#the index is reset to the default numeric index.
```
